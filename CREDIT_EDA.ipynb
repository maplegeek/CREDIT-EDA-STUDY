```python
#Diving our merged data set into - The ones who have payment difficulties and the ones who do not have any payment difficulties. 
target0=merged[merged['TARGET']==0] #The ones who do not have payment difficulties
target1=merged[merged['TARGET']==1] # The ones who have payment difficulties 
```

```python
def pie(name):
    plt.subplots(1,2,figsize=[18,10])
    plt.subplot(1,2,1)
    plt.title('Clients without payment difficulties')
    target0[name].value_counts().plot.pie(autopct='%0.1f%%')
    plt.subplot(1,2,2)
    plt.title('Clients with payment difficulties')
    target1[name].value_counts().plot.pie(autopct='%0.1f%%')
    plt.show()
```

```python
pie('NAME_CONTRACT_STATUS')
```

#### Clients without payment difficulties are more likely to get their loans approved.
#### Also, their refusal percentage is lower compared to clients with payment difficulties.

## Analysing Approved and Rejected Data from previous application.

```python
# Dividing data into approved and rejected loans.
rejected=merged[merged['NAME_CONTRACT_STATUS']=='Refused']
approved=merged[merged['NAME_CONTRACT_STATUS']=='Approved']
```

## Univariate Analysis

```python
def RA_pie(name):
    plt.subplots(1,2,figsize=[16,8])
    plt.subplot(1,2,1)
    approved[name].value_counts().plot.pie(autopct='%0.1f%%')
    plt.title('Approved Application')
    plt.subplot(1,2,2)
    rejected[name].value_counts().plot.pie(autopct='%0.1f%%')
    plt.title('Rejected Application')
    plt.ylabel(None)
    plt.show()
```

```python
RA_pie('TARGET')
```

#### For Approved applications (0), clients who had payment difficulties is only 8.1%
